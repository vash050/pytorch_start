```python
import pandas as pd
data = pd.read_csv("/content/adult-census-income-dataset/adult.csv")
```

```python
data.head
```

```python
data_inputs = data.iloc[:, [0,1,2,3,4,5,6,7,8]]
```

```python
import numpy as np

data_np_array = data_inputs.to_numpy('float32')
```

```python
data_targets = data.iloc[:, [9]]
```

```python
targets_np_array = data_targets.to_numpy(dtype='float32')
```

```python
data_np_array
```

```python
targets_np_array.dtype
```

```python
import torch
```

```python
inputs = torch.from_numpy(data_np_array)
targets = torch.from_numpy(targets_np_array)
```

```python
inputs
```

```python
inputs.shape
```

```python
w = torch.randn(1, 9, requires_grad=True)
b = torch.randn(1, requires_grad=True)
print(w)
print(b)
```

```python
def model(x):
  return x @ w.t() + b
```

```python
preds = model(inputs)
print(preds)
```

```python
print(targets)
```

```python
diff = preds - targets
torch.sum(diff * diff) /diff.numel()
```

```python
def mse(t1, t2):
  diff = t1 -t2
  return torch.sum(diff * diff) / diff.numel()
```

```python
loss = mse(preds, targets)
print(loss)
```

```python
loss.backward()
```

```python
print(w)
print(w.grad)
```

```python
print(b)
print(b.grad)
```

```python
with torch.no_grad():
  w -= w.grad * 1e-5
  b -= b.grad * 1e-5
```

```python
w, b
```

```python
loss = mse(model(inputs), targets)
print(loss)
```

```python

w.grad.zero_()
b.grad.zero_()
print(w.grad)
print(b.grad)
```

```python
for i in range(100):
  preds = model(inputs)
  loss = mse(preds, targets)
  loss.backward()
  with torch.no_grad():
    w -= w.grad * 1e-5
    b -= b.grad * 1e-5
    w.grad.zero_()
    b.grad.zero_()
```
